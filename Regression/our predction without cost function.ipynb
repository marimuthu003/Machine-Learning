##### Learn to implement the model f(w,b) for linear regression with one variable

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-deep')
```

```python

# x_train is the input variable(size in 100 square feet)
# y_train is the output variable(price in 1000 dollars)
x_train=np.array([1.0,2.0])
y_train=np.array([300,500])
```

```python
# m - no of training examples
m=len(x_train)
print(f"no of training examples is {m}")
```

```python
# training example x^i,y^i
i=0 # take this for also 1
x_i=x_train[i]
y_i=y_train[i]

print(f"training example : x^{i},y^{i} = {x_i},{y_i}")
```

```python
plt.scatter(x=x_train,y=y_train,marker='x',c='r')
plt.title('House Price Prediction')
plt.xlabel('size of feet^2 in 100')
plt.ylabel('price in 1000 dollars')
plt.show()
```

```python
w=100
b=100
print(f"w:{w}")
print(f"b:{b}")
```

```python
def calculate_model_output(w,b,x):
    m=x.shape
    f_wb=np.zeros(m)
    for i in range(len(x)):
        f_wb[i]=w * x[i]+b
    return f_wb
```

```python
tmp_f_wb=calculate_model_output(w,b,x_train)

plt.plot(x_train,tmp_f_wb,c='b',label='our prediction')

plt.scatter(x=x_train,y=y_train,marker='x',c='r',label='Actual value')
plt.title('House Price Prediction')
plt.xlabel('size of feet^2 in 100')
plt.ylabel('price in 1000 dollars')
plt.legend()
plt.show()
```

#### As you can see w=100 and b=100 does not result in a line that fits our data
